```python
# Проверим, влияет ли уровень экологической обстановки в районе 
# на цену квартиры. Постройте коробчатую диаграмму цен на квартиры 
# (price_doc) в зависимости от уровня экологической обстановки в 
# районе (ecology).
fig = plt.figure(figsize=(10, 7))
boxplot = sns.boxplot(
    data=sber_data,
    x='price_doc',
    y='ecology',
    width=0.95
)
boxplot.set_title('Распределение цены по уровню экологии');
boxplot.set_ylabel('Уровень экологии');
boxplot.set_xlabel('Цены на квартиры');
boxplot.grid()
```

```python
# Постройте диаграмму рассеяния, которая покажет, как цена на квартиру 
# (price_doc) связана с расстоянием до центра Москвы (kremlin_km).

# matplotlib
fig = plt.figure(figsize=(8, 4))
axes = fig.add_axes([0, 0, 1, 1])
axes.scatter(
    x=sber_data['price_doc'], 
    y=sber_data['kremlin_km'], 
    s=10,
    marker='h',
    c='purple'
)
axes.set_title('Зависимость цены от расстояния до центра Москвы', fontsize=16)
axes.set_xlabel('Цена на квартиру')
axes.set_ylabel('Расстояние до центра Москвы')
axes.xaxis.set_tick_params(rotation=30)
axes.grid()

# seaborn
fig = plt.figure(figsize=(8, 5))
scatterplot = sns.scatterplot(
    data=sber_data,
    x='price_doc',
    y='kremlin_km',
    hue='kremlin_km',
    s=50,
    size='price_doc',
    sizes=(20, 300)
)
scatterplot.set_title('Зависимость цены от расстояния до центра Москвы', fontsize=12)
scatterplot.set_xlabel('Цена на квартиру')
scatterplot.set_ylabel('Расстояние до центра Москвы');
```
